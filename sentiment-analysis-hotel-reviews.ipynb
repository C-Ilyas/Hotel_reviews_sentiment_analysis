```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import nltk
import glob
```

```python
df=pd.read_csv('/kaggle/input/hotel-reviews/Datafiniti_Hotel_Reviews.csv')
df.head()
```

```python
df=df.head(1000)
print(df.shape)
```

```python
ax=df['reviews.rating'].value_counts().sort_index().plot(kind='bar',title='score reviews',figsize=(10,5))
ax.set_xlabel('reviews stars')
plt.show()
```

```python
example=df['reviews.text'][50]
example
```

```python
tokens=nltk.word_tokenize(example)
tokens
```

```python
tagged=nltk.pos_tag(tokens)
tagged[:10]
```

```python
entities=nltk.chunk.ne_chunk(tagged)
entities.pprint()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sia=SentimentIntensityAnalyzer()
```

```python
res={}
for i , row in tqdm(df.iterrows(),total=len(df)):
    id=row['id']
    text=row['reviews.text']
    res[id]=sia.polarity_scores(text)
    
```

```python
vaders=pd.DataFrame(res).T
vaders=vaders.reset_index().rename(columns={'index':'id'})
vaders=vaders.merge(df,how='left')
vaders.head()
```

```python
ax=sns.barplot(data=vaders, x='reviews.rating',y='compound')
ax.set_title('compound reviews by star reviews')
plt.show()
```

```python
fig,axs=plt.subplots(1,3,figsize=(15,3))
sns.barplot(data=vaders,x='reviews.rating',y='pos',ax=axs[0])
sns.barplot(data=vaders,x='reviews.rating',y='neu',ax=axs[1])
sns.barplot(data=vaders,x='reviews.rating',y='neg',ax=axs[2])
axs[0].set_title('positive reviews')
axs[1].set_title('neutral reviews')
axs[2].set_title('negative reviews')
plt.tight_layout()
plt.show()
```
